1. Write a Python program to draw a line with suitable label in the x axis, y axis and a title. The code snippet gives the output shown in the following screenshot:

4. Write a Python program to plot two or more lines with legends, different widths and colors. The code snippet gives the output shown in the following screenshot:

```python
import matplotlib.pyplot as plt
import numpy as np

x1=np.array([10,20,30])
y1=np.array([20,40,10])

plt.plot(x1,y1,  linewidth =3,color="b")

x2=np.array([10,20,30])
y2=np.array([40,10,30])

plt.plot(x2,y2, linewidth = 5,color="r")
plt.title("Sample graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend(["line1-width-3","line2-width-5"])
plt.show()
```

6. Write a Python programming to display a bar chart of the popularity of programming Languages.

```python
import matplotlib.pyplot as plt
import numpy as np

x=(['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++'])
y=([22.2, 17.6, 8.8, 8, 7.7, 6.7])

plt.bar(x,y)
plt.grid()
plt.title("Popularity of programming language \n worldwide,oct 2017 compared to a year ago")
plt.xlabel("languages")
plt.ylabel("popularity")
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

df=pd.read_csv('company_sales_data.csv')
profit=df["total_profit"].to_list()
month=df["month_number"].to_list()
plt.plot(month,profit,linestyle='--',linewidth=3,color='r',marker='o',markerfacecolor='k')
plt.title("company sales data of last year")
plt.xlabel("Month number")
plt.ylabel("solid unites number")
plt.legend(title = "profit data of last year",loc="lower right")
plt.show()
```

Exercise 3: Read all product sales data and show it using a multiline plot

 5: Read face cream and facewash product sales data and show it using the bar chart
The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

The bar chart should look like this.



```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("company_sales_data.csv")
monthList  = df ['month_number'].tolist()
faceCremSalesData   = df ['facecream'].tolist()
faceWashSalesData   = df ['facewash'].tolist()

plt.bar([a-0.25 for a in monthList], faceCremSalesData, width= 0.25, label = 'Face Cream sales data', align='edge')
plt.bar([a+0.25 for a in monthList], faceWashSalesData, width= -0.25, label = 'Face Wash sales data', align='edge')
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.legend(loc='upper left')
plt.title(' Sales data')

plt.xticks(monthList)
plt.grid(True, linewidth= 1, linestyle="--")
plt.title('Facewash and facecream sales data')
plt.show()
```

Exercise 7: Read the total profit of each month and show it using the histogram to see the most common profit ranges

```python
import matplotlib.pyplot as plt
import pandas as pd

df=pd.read_csv("company_sales_data.csv")

profit=df['total_profit'].to_list()

profit_range = [150000, 175000, 200000, 225000, 250000, 300000, 350000]

plt.hist(profit,profit_range, label="profit data")

plt.xlabel('Profit range in Dollar')
plt.ylabel('Actual profit in Dollar')

plt.title(' Profit Data')
plt.legend(loc="upper left")

plt.show()
```

Exercise 8: Calculate total sale data for last year for each product and show it using a Pie chart

```python
import matplotlib.pyplot as plt
import pandas as pd

df=pd.read_csv("company_sales_data.csv")

total_sale=(df['facecream'].sum(),df['facewash'].sum(),df['toothpaste'].sum(),df['bathingsoap'].sum(),df['shampoo'].sum(),df['moisturizer'].sum())

plt.pie(total_sale,labels=['facecream','facewash','toothpaste','bathingsoap','shampoo','moisturizer'],autopct='%1.1f%%')

plt.legend(loc="lower right")

plt.title("Sales Data")
plt.show()
```

Exercise 9: Read Bathing soap facewash of all months and display it using the Subplot

```python
import matplotlib.pyplot as plt
import pandas as pd

df=pd.read_csv("company_sales_data.csv")

month=df["month_number"].to_list()
facewashsd=df["facewash"].to_list()

plt.subplot(2,1,1)
plt.title("Sales Data of Face Wash")
plt.plot(month,facewashsd, linewidth=3,marker='o',color='r')

bathingsoapsd = df ['bathingsoap'].tolist()
plt.subplot(2,1,2)
plt.title("Sales Data of Bathing Soap")
plt.plot(month,bathingsoapsd,linewidth=3,marker='o')

plt.show()
```

Exercise Question 10: Read all product sales data and show it using the stack plot

```python
import matplotlib.pyplot as plt
import pandas as pd

df=pd.read_csv("company_sales_data.csv")
month=df["month_number"].to_list()
facecreamsd=df["facecream"].to_list()
facewashsd=df["facewash"].to_list()
toothpastesd = df ['toothpaste'].tolist()
bathingsoapsd   = df ['bathingsoap'].tolist()
shampoosd   = df ['shampoo'].tolist()
moisturizersd = df ['moisturizer'].tolist()

label=['facewash','facecream','toothpaste','bathingsoap','shampoo','moisturizer']
plt.stackplot(month,facecreamsd,facewashsd,toothpastesd,bathingsoapsd,shampoosd,moisturizersd)
plt.title("All product sales data using stack plot")
plt.legend(loc='upper left', labels=label)
plt.show()
```
